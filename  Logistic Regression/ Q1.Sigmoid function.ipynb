### â€¢ Find the values of the sigmoid function for values of x from -10.0 to 10.0 in steps of 0.1. Plot these values on a graph. 


```python
import math
import matplotlib.pyplot as plt
```

```python
x=[]
y=[]

for z in range(-10,10):
    x.append(z)
    y.append(1/(1+math.exp(-z)))
print(y)
```

```python
plt.scatter(x,y)
plt.plot(x,y)
plt.title("Sigmoid function for value from -10 to 10")
plt.xlabel("x-values")
plt.ylabel("y_values")
plt.savefig("sigmoid.png")
plt.show()
```
