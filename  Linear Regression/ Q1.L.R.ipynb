```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
```

```python
x = np.array([[1, 2, 3, 4, 5, 6]]).reshape(-1,1)
y = np.array([6,1,9,5,17,12])
x

```

```python
model = LinearRegression()
model.fit(x,y)
print("Intercept: ", model.intercept_)
print("Slope: ",model.coef_)
```

```python
y_predict = model.predict(x)
Root_mean_Square_error = np.sqrt(mean_squared_error(y,y_predict))
R2 = model.score(x,y)
print("Model_Root_mean_Square_Error: ", Root_mean_Square_error)
print("R2: ",R2)
```

```python
plt.scatter(x, y, color='black',label = 'actual')
plt.plot(x, y_predict, color='blue', linewidth=3,label='predict')
plt.title("ACTUAL AND PREDICTED GRAPH")
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend(loc = 'upper left')
plt.savefig('predictactual.png',dpi=300)
plt.show()
```
