```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

#### Define the Data

## Simple linear regression Example(one feature x)

```python
x = np.array([1, 2, 3, 4, 5, 6]).reshape(-1, 1)
y = np.array([6,1,9,5,17,12])
print("x.shape ", x.shape)
print("y.shape ", y.shape)
model = LinearRegression()
model.fit(x, y)
print('intercept:', model.intercept_)
print('slope:', model.coef_)
y_hat = model.predict(x)
print('predicted response:', y_hat, sep='\n')
plt.scatter(x, y, color='black')
plt.plot(x, y_hat, color='blue', linewidth=3)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
x = np.array([1, 2, 3, 4, 5, 6]).reshape(-1, 1)
y = np.array([6,1,9,5,17,12])
print("x.shape ", x.shape)
print("y.shape ", y.shape)
```

#### Build the Model

```python
model = LinearRegression()
model.fit(x, y)
print('intercept:', model.intercept_)
print('slope:', model.coef_)
print('R^2: ', model.score(x,y))
```

#### This is not a great model. Coefficient of Determinaton is 0.49. Only about half the variation in y is explained by the model.

#### Prediction making 

```python
y_hat = model.predict(x)
print('predicted response:', y_hat, sep='\n')
model.predict([[2]])
```

#### Draw a graph

```python
plt.scatter(x, y, color='black', label = 'actual')
plt.scatter(x, y_hat, color = 'blue', label = 'Predict')
plt.plot(x, y_hat, color='blue', linewidth=3)
plt.title("Predicted and actual value of y")
plt.xlabel('x_Axis')
plt.ylabel('y_Axis')
plt.legend(loc = 'upper left')
plt.show()
```
