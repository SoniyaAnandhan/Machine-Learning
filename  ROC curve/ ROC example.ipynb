```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
rocData = pd.read_csv("rocExample.csv")
print(rocData.head())
N = 10
P = 10
total = 20
tprList = list()
fprList = list()
tList = list()
```

```python
for i in range(total) :
 predicted = np.zeros(20)
 TPVector = np.zeros(20)
 FPVector = np.zeros(20)

 threshold = rocData.score[i]
 tList.append(threshold)
 for j in range(total) :
     if rocData.score[j] >= threshold :
         predicted[j] = 1
 for j in range(total) :
     if rocData.actual[j] == 'p' and predicted[j] == 1 :
         TPVector[j] = 1
     if rocData.actual[j] == 'n' and predicted[j] == 0 :
         FPVector[j] = 1

 TPR = sum(TPVector) / P
 FPR = sum(FPVector) / N

 TP = sum(TPVector)
 TN = sum(FPVector)

 accuracy = (TP+TN)/total
 
 fprList.append(FPR)
 tprList.append(TPR)
 print("Throshold: ", threshold, "Accuracy:",accuracy)
```

```python
plt.scatter(fprList, tprList, color='red')
plt.plot(fprList, tprList, color='black')
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.title('ROC Curve')
plt.savefig('roc1.png')
plt.show()
# plt.figure()
fig, ax = plt.subplots()
for i, txt in enumerate(tList):
 ax.annotate(txt, (fprList[i], tprList[i]))
 plt.scatter(fprList[i], tprList[i], color='black')
plt.plot(fprList, tprList, color='black')
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.title('ROC Curve')
plt.savefig('roc2.png')
plt.show()
```
